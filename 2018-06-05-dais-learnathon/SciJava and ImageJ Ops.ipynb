# ImageJ Ops

* A framework for reusable image processing operations
* Extends Java's idea of "write once, run anywhere" to image processing algorithms
* Part of [ImageJ2](https://imagej.net/ImageJ2)

## https://imagej.net/Ops
## https://imagej.github.io/tutorials

```scijava
#@ ImageJ ij
clown = ij.io().open("https://imagej.net/images/clown.jpg")
```

```scijava
slice = ij.op().transform().hyperSliceView(clown, 2, 0)
```

```scijava
ij.op().filter().gauss(slice, [20, 2] as double[])
```
